Multiclass classification On Airlines Delay Dataset.

```python
import pandas as pd     # Importing important laibraries for the project
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils #one hot encoding
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
df_air=pd.read_csv("/content/airlines_delay.csv")    # reading the csv file using read_csv function
```

```python
df=df_air.dropna(axis=0)  # removing NAN rows using dropna function. 
```

```python
df.isna().sum()   # NAN are removed.
```

```python
x=df.iloc[:,:-1].values  # converting dataframe into numpy array
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
le=LabelEncoder() #  converting categorical data into continuous data using Labelencoder for column Airline.
x[:,3]=le.fit_transform(x[:,3])
```

```python
le=LabelEncoder() # converting categorical data into continuous data using Labelencoder for column AirportFrom.
x[:,4]=le.fit_transform(x[:,4])
```

```python
le=LabelEncoder() # converting categorical data into continuous data using Labelencoder for column  AirportTo.
x[:,5]=le.fit_transform(x[:,5])
```

```python
x  # after encoding values of X
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2, random_state=1) # Splitting the data using train_test_split.
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc=StandardScaler() # Scaling the data using StandardScaler.
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
ann=Sequential()   #step 1: Initialize ANN model
```

```python
# step 2: Adding layers
ann.add(Dense(units=6,activation="relu"))#i/p layer
ann.add(Dense(units=4,activation="relu"))#hidden layer
ann.add(Dense(units=1,activation="softmax"))#output layer
```

```python
# Step 3: Establish connection
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
# Step 4: Train model
ann.fit(xtrain,ytrain,epochs=100)
```

```python
# Step 5: Predicting actual and predicted values..
pred=ann.predict(xtest)
air_length=len(pred)
```

```python
ylabel=np.argmax(ytest)#actual
predlabel=np.argmax(pred)#predicted
```

```python
ylabel
```

```python
predlabel
```

```python
ac=np.sum(ylabel==predlabel)/air_length*100
print(f"Accuracy:{ac}")
```

```python
def airmodel():  # creating function named airmodel for the model.
  ann=Sequential()

  ann.add(Dense(units=6,activation="relu"))#i/p layer
  ann.add(Dense(units=4,activation="relu"))#hidden layer
  ann.add(Dense(units=1,activation="softmax"))#output layer

  ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])

  ann.fit(xtrain,ytrain,epochs=100)
  return ann
```

```python
#  Evaluating the model.
estimator = KerasClassifier(build_fn= airmodel, epochs=200, batch_size=5, verbose=0 )
cvs = cross_val_score(estimator, x,y, cv=10)
print(f"Avg Accuracy -: {cvs.mean()*100}\nSTD : {cvs.std()}")
```
